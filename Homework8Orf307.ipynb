```python
import cvxpy as cp
import numpy as np

# Define variables (continuous)
x1 = cp.Variable()
x2 = cp.Variable()

# Objective function
objective = cp.Minimize(-x1 - 2*x2)

# Constraints
constraints = [
    -3*x1 + 4*x2 <= 4,
    3*x1 + 2*x2 <= 11,
    2*x1 - x2 <= 5,
    x1 >= 0,
    x2 >= 0
]

# Define and solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

print("Optimal value (Linear Programming Relaxation):", problem.value)
print("x1: ", x1.value)
print("x2:", x2.value)
```
